```python
import os
import sys
import re
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException
```

```python
# Configurations.
debug = False
download_path = os.path.join(os.getcwd(), "./download")
download_path = os.path.abspath(download_path)
homeurl = "https://dl.acm.org/journal/imwut/editorial-board"

# Load data if exported.
# df = pd.read_csv("./IMWUT.csv")
# df.head(2)
```

```python
# Start a webdriver instance.
options = webdriver.ChromeOptions()
prefs = {}
prefs["profile.default_content_settings.popups"] = 0
prefs["download.default_directory"] = download_path
options.add_experimental_option("prefs", prefs)

driver = webdriver.Chrome("./chromedriver.exe", options=options)
```

```python
# Load page.
driver.get(homeurl)
```

```python
# Find all metadata.
all_metadata_elements = driver.find_elements_by_xpath("//div[@class='item-meta__info']")

# Iterating names. 
all_editors = []
for metadata_element in all_metadata_elements:
    editor_name = metadata_element.find_element_by_xpath(".//h4").get_attribute('innerHTML')
    editor_institute = metadata_element.find_element_by_xpath(".//p").get_attribute('innerHTML')
    editor_country = metadata_element.find_element_by_xpath(".//span").get_attribute('innerHTML')
    all_editors.append({
        "editor_name": editor_name,
        "editor_institute": editor_institute,
        "editor_country": editor_country
    })

df = pd.DataFrame.from_dict(all_editors)
df
```

```python
# Export.
df.to_csv("./exports/imwut_editors.csv", index=False)
```

```python
# Show hist.
df["editor_country"].value_counts().plot.barh()
plt.show()
df["editor_institute"].value_counts().plot.barh(figsize=(5, 20))
```
