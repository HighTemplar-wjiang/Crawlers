```python
# Collating bib files.
all_bib_files = os.listdir(download_path)

all_bib_entries = ""
for bib_file in all_bib_files:
    if not bib_file.endswith(".bib"):
        continue
    
    print(bib_file)
    with open(os.path.join(download_path, bib_file), "r", encoding="utf-8") as f:
        all_bib_entries += f.read()

# Save.
with open("export.bib", "w", encoding="utf-8") as f:
    f.write(all_bib_entries)
```
